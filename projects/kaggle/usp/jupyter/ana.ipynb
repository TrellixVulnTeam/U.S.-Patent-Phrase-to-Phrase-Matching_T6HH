```python
def count_merge(l):
  m = defaultdict(int)
  for key in l:
    m[key] += 1
  res = []
  for key in m:
    res.append(f'{key} {m[key]}')
  return ';'.join(res)
dfc['contexts'] = dfc.contexts.apply(count_merge)
dfc
```
